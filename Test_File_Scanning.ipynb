# Now For All Pages

```python
with open('Test_File.pdf','rb') as fid:
    pdfReader = PyPDF2.PdfFileReader(fid) 
    print(pdfReader.numPages)
    alltext=[pdfReader.getPage(page).extractText() for page in range(pdfReader.numPages)]  
```

```python
data=[]

a_1=7
a_2=14

lines=str(alltext).split('\\n')
        
while a_2 <= len(lines):

    a=lines[a_1:a_2]

    a_1+=7
    a_2+=7


    dic={}

    for key,value in zip(['Row','Col_1','Col_2','Col_3','Col_4','Col_5','Col_6'],a):
        dic[key]=value

    data.append(dic)
data=pd.DataFrame(data)
data[:5]
```
